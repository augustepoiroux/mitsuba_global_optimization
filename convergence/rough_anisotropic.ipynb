```python
import mitsuba as mi

mi.set_variant("cuda_ad_rgb")
import drjit as dr
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
import numpy as np

import os
import sys
ROOT_DIR = os.path.dirname(os.getcwd())
sys.path.append(ROOT_DIR)

from utils.utils import image_to_bm, to_float, unidim_to_bm, rel_l1_loss, mse, LOSS_FNS
from utils.plot import (
    plot_loss_linear_interp_1d,
    plot_loss_bilinear_interp_2d,
    plot_losses,
)

mi.Thread.thread().logger().set_log_level(mi.LogLevel.Error)
```

```python
def reset_scene():
    return mi.load_file(
        os.path.join(
            ROOT_DIR, f"scenes/rough_anisotropic_optimization.xml"
        )
    )
```

```python
def f(x, alpha, beta):
    return 4 * (alpha -beta) * x * (x - 1) + alpha

def init_scene_params(theta):
    scene = reset_scene()
    params = mi.traverse(scene)
    params["plane.bsdf.alpha_u.value"] = f(theta, 0.3, 0.05)
    params["plane.bsdf.alpha_v.value"] = f(theta, 0.05, 0.3)
    params["red_right.intensity.value"] *= 0.2 * (1 - theta) + theta
    params["red_left.intensity.value"] *= (1 - theta) + theta * 0.2
    params.update()
    return scene, params
```

# Reference scenes rendering

```python
scene_ref, param_ref = init_scene_params(0.0)

# Render reference images
img_ref = mi.render(scene_ref, seed=0, spp=64)
img_ref_bm = image_to_bm(img_ref)

# Plot
plt.imshow(img_ref_bm)
```

```python
scene, params = init_scene_params(0.5)

params.update()

# Render images
image = mi.render(scene, seed=0, spp=64)
image_bm = image_to_bm(image)

# Plot
plt.imshow(image_bm)
```

```python
n_samples = 30
spp = 32

alphas = np.linspace(0, 1, n_samples)
losses = {key: [] for key in LOSS_FNS.keys()}
for i, alpha in enumerate(alphas):
    scene, params = init_scene_params(alpha)
    
    image = mi.render(scene, params, seed=i, spp=spp)
    for key in LOSS_FNS.keys():
        losses[key].append(LOSS_FNS[key](image, img_ref))

fig, axs = plt.subplots(len(losses), 1, sharex=True)
for i, key in enumerate(losses.keys()):
    axs[i].plot(alphas, losses[key])
    axs[i].set_ylabel(key)
plt.show()
```

```python
opt = mi.ad.Adam(lr=0.02)

# Our latent variable
theta_init_value = 0.6
opt_key = "theta"
opt[opt_key] = mi.Float(theta_init_value)

_, param_ref = init_scene_params(theta_init_value)
scene, params = init_scene_params(theta_init_value)

def update_scene():
    theta = dr.clamp(opt[opt_key], 0, 1)
    params["plane.bsdf.alpha_u.value"] = f(theta, 0.3, 0.05)
    params["plane.bsdf.alpha_v.value"] = f(theta, 0.05, 0.3)
    params["red_right.intensity.value"] = param_ref["red_right.intensity.value"] * (0.2 * (1 - theta) + theta)
    params["red_left.intensity.value"] = param_ref["red_left.intensity.value"] * ((1 - theta) + theta * 0.2)
    params.update()

update_scene()
```

```python
image_init = mi.render(scene, spp=128)
image_to_bm(image_init)
```

```python
iteration_count = 100
errors = []
for it in range(1, iteration_count+1):
    # Perform a (noisy) differentiable rendering of the scene
    image = mi.render(scene, params, spp=4)

    # Evaluate the objective function from the current rendered image
    loss = mse(image, img_ref)

    # Backpropagate through the rendering process
    dr.backward(loss)

    # Optimizer: take a gradient descent step
    opt.step()
    opt[opt_key] = dr.clamp(opt[opt_key], 0, 1)

    # Update the scene state to the new optimized values
    update_scene()

    print(f"Iteration {it:02d}: loss={to_float(loss):.3f}\ttheta={to_float(opt[opt_key]):.3f}", end='\r')
print('\nOptimization complete.')
```
